## Serie de Fourier (exponencial, trigonométrica y compacta)

### **Semejanzas**
- Las tres permiten representar una **señal periódica** como la suma de muchas ondas más simples (armónicos) de distintas frecuencias.  
- En todas se puede observar **qué frecuencias componen la señal** y con qué **amplitud y fase** aparecen.  
- Son **equivalentes entre sí**: expresan la misma información, pero con diferente forma matemática.

### **Diferencias**

**Forma trigonométrica:**  
Utiliza funciones reales, senos y cosenos:
$$
x(t) = \frac{a_0}{2} + \sum_{n=1}^{\infty} [a_n \cos(n\omega_0 t) + b_n \sin(n\omega_0 t)]
$$
Es la más intuitiva y visual, ya que muestra cuánto “coseno” y cuánto “seno” contiene la señal.  
Los coeficientes \(a_n\) y \(b_n\) indican la **amplitud** de cada componente.



**Forma exponencial o compleja:**  
Emplea exponenciales complejas en lugar de senos y cosenos:
$$
x(t) = \sum_{k=-\infty}^{\infty} C_k e^{j k \omega_0 t}
$$
Es más conveniente para cálculos matemáticos y análisis en ingeniería.  
Los coeficientes \(C_k\) son números complejos que combinan la información de \(a_n\) y \(b_n\) en una sola expresión.



**Forma compacta o fasorial:**  
Muestra la señal como la suma de ondas cosenoidales con diferentes amplitudes y fases:
$$
x(t) = C_0 + \sum_{n=1}^{\infty} C_n \cos(n\omega_0 t + \theta_n)
$$
La componente \(n=1\) corresponde a la frecuencia fundamental, mientras que las \(n>1\) representan los **armónicos** que definen el detalle y la forma de la señal.  
Esta forma permite visualizar claramente la **amplitud y fase** de cada armónico.


**Observación:**  
Las tres versiones describen el mismo fenómeno: cómo una señal periódica se puede construir a partir de ondas más simples.  
La forma trigonométrica es la más fácil de interpretar visualmente, mientras que las formas exponencial y compacta resultan más prácticas para el análisis y los cálculos.



## Transformada de Fourier (TF), DTFT y DFT

### **Semejanzas**
- Todas analizan una señal en el **dominio de la frecuencia**, mostrando de qué frecuencias está compuesta.  
- Buscan la misma relación entre **tiempo y frecuencia**, solo que en distintos contextos (continuo o discreto).  
- Están **matemáticamente relacionadas**: la DTFT y la DFT son versiones discretas o muestreadas de la Transformada de Fourier continua.  
- En todos los casos existe una **transformada directa** y una **transformada inversa** que permiten pasar entre ambos dominios.


### **Diferencias**

**Transformada de Fourier (TF):**  
Se aplica a señales **continuas** en el tiempo, generalmente **aperiódicas**.  
Su resultado es un **espectro continuo** en frecuencia:
$$
X(\omega) = \int_{-\infty}^{\infty} x(t)e^{-j\omega t}dt
$$
Transformada inversa:
$$
x(t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} X(\omega)e^{j\omega t}d\omega
$$
Permite analizar cómo se distribuye la energía de una señal continua a lo largo de todas las frecuencias (por ejemplo, señales de audio analógicas).


**Transformada de Fourier en Tiempo Discreto (DTFT):**  
Se aplica a señales **discretas** (muestreadas en el tiempo) pero de **duración infinita**.  
El resultado es un **espectro continuo** en frecuencia, aunque **periódico** con periodo \(2\pi\):
$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j\omega n}
$$
Inversa:
$$
x[n] = \frac{1}{2\pi} \int_{-\pi}^{\pi} X(e^{j\omega}) e^{j\omega n} d\omega
$$
La periodicidad del espectro surge porque al muestrear una señal en el tiempo, su espectro se repite en frecuencia.  
Es muy usada en el análisis de **sistemas digitales**.


**Transformada Discreta de Fourier (DFT):**  
Se utiliza con señales **discretas y de duración finita** (N muestras).  
El resultado es un **espectro discreto y periódico**, compuesto por N puntos:
$$
X[k] = \sum_{n=0}^{N-1} x[n] e^{-j \frac{2\pi}{N}kn}, \qquad k=0,1,\dots,N-1
$$
Inversa:
$$
x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] e^{j \frac{2\pi}{N}kn}
$$
La DFT es la versión **numérica y práctica** de la transformada, implementable en computador.  
Su cálculo eficiente se realiza con el algoritmo **FFT (Fast Fourier Transform)**.


### **Relación entre señal y espectro**

| Tipo de señal | Tipo de espectro | Transformada asociada |
|----------------|------------------|------------------------|
| Señal continua | Espectro continuo | Transformada de Fourier (TF) |
| Señal discreta (infinita) | Espectro continuo y periódico | DTFT |
| Señal discreta y finita | Espectro discreto y periódico | DFT |

- **Muestrear la señal** en el tiempo genera **periodicidad** en el espectro.  
- **Hacer la señal finita** (tomar solo N muestras) produce un espectro **discreto** con un número finito de componentes.


**Observación:**  
Las tres transformadas representan el mismo principio de análisis de frecuencia, pero en distintos escenarios:
- La **TF** se usa en señales analógicas (tiempo continuo).  
- La **DTFT** analiza señales digitales infinitas.  
- La **DFT** es la versión práctica usada en procesadores y software, calculada mediante la FFT.  
Todas buscan descomponer una señal en sus **frecuencias básicas**, variando únicamente si la señal y el espectro son continuos o discretos.


## Transformada Rápida de Fourier (FFT)

### **Qué es y para qué sirve**
La FFT no es una nueva transformada, sino un **algoritmo optimizado** para calcular la DFT mucho más rápido.  
Permite reducir el número de operaciones de \(O(N^2)\) a \(O(N\log N)\), lo que la hace esencial en aplicaciones donde se requiere procesar señales en tiempo real.

### **Semejanzas (FFT vs DFT)**
- Ambas calculan los mismos valores \(X[k]\) para una secuencia \(x[n]\) de N muestras.  
- Usan exactamente la misma fórmula base:
  $$
  X[k] = \sum_{n=0}^{N-1} x[n] e^{-j \frac{2\pi}{N}kn}
  $$
- La FFT solo cambia la **forma de cálculo**, no el resultado.

### **Diferencias**
**DFT (cálculo directo):**
- Evalúa la fórmula de manera literal.  
- Requiere aproximadamente \(N^2\) multiplicaciones complejas.  
- Se vuelve ineficiente para valores grandes de N.

**FFT (Fast Fourier Transform):**
- Utiliza una estrategia **recursiva** para dividir la DFT grande en DFTs más pequeñas.  
- Reduce el costo a \(N\log_2 N\) operaciones.  
- Hace posible aplicar la DFT en tiempo real y en hardware limitado.



### **Idea básica de la FFT**

1. **Descomposición recursiva:**  
   Divide la DFT de tamaño \(N\) en dos DFTs de tamaño \(N/2\), una para los índices pares y otra para los impares.

2. **Reutilización de cálculos:**  
   Las DFT pequeñas se combinan mediante factores complejos llamados **twiddle factors**:  
   $$
   W_N^k = e^{-j \frac{2\pi k}{N}}
   $$

3. **Combinación tipo butterfly:**  
   Los resultados parciales se mezclan con operaciones simples de suma y resta, formando estructuras llamadas **butterflies**.

4. **Recursividad:**  
   Este proceso se repite hasta llegar a DFTs de tamaño 1.


### **Importancia práctica**
Gracias a la FFT, el análisis espectral y el procesamiento digital de señales se volvieron viables en tiempo real.  
Hoy es una herramienta clave en **audio, imagen, comunicaciones, radar, biomedicina y electrónica digital**, entre muchas otras áreas.

# Modulación por Amplitud con Detección Coherente

## ¿En qué consiste?

La **modulación por amplitud (AM)** es una técnica de transmisión en la que la **amplitud de una señal portadora** de alta frecuencia varía de acuerdo con la amplitud instantánea de una **señal mensaje**, que contiene la información que se desea enviar.

La **detección coherente** (o **demodulación síncrona**) es el proceso inverso: se encarga de **recuperar la señal original** a partir de la señal modulada recibida. Para lograrlo, el receptor genera localmente una **portadora de referencia** que tiene **la misma frecuencia y fase** que la portadora original usada en el transmisor.  
Al multiplicar esta portadora local por la señal recibida y aplicar luego un **filtro pasa bajos**, se eliminan las componentes de alta frecuencia y se obtiene nuevamente la señal mensaje.



## Principio de funcionamiento

Para realizar la modulación se utilizan dos señales principales:

### 1. Señal portadora

Es una onda de alta frecuencia, generalmente senoidal o cosenoidal, expresada como:

$$
c(t) = A_c \sin(2\pi f_c t)
$$

donde:

- \( A_c \): amplitud de la portadora  
- \( f_c \): frecuencia de la portadora  


### 2. Señal mensaje

Es la señal que contiene la información (de baja frecuencia). Usualmente se representa por una senoidal:

\[
m(t) = A_m \cos(2\pi f_m t)
\]

donde:

- \( A_m \): amplitud de la señal mensaje  
- \( f_m \): frecuencia del mensaje  



### 3. Índice de modulación

El grado de variación de la amplitud de la portadora se mide con el **índice de modulación**, definido como:

$$I_m = \frac{\text{Peak}\{|m(t)|\}}{A_c}$$

Este parámetro debe cumplir \( 0 \leq I_m \leq 1 \) para evitar **sobremodulación**, que provoca distorsión en la señal transmitida.



## Señal modulada en AM

La señal modulada se obtiene multiplicando la portadora por un factor dependiente de la señal mensaje:

$$y(t) = [1 + \frac{m(t)}{A_c}] \, c(t)
$$

Sustituyendo \( c(t) = A_c \sin(2\pi f_c t) \), se tiene:

$$
y(t) = A_c [1 + \frac{A_m}{A_c} \cos(2\pi f_m t)] \sin(2\pi f_c t)
$$

Usando el índice de modulación \( I_m = \frac{A_m}{A_c} \):

$$
y(t) = A_c [1 + I_m \cos(2\pi f_m t)] \sin(2\pi f_c t)
$$



## Características principales

- La **detección coherente** exige **sincronización precisa** entre la portadora del transmisor y la generada localmente en el receptor (misma frecuencia y fase).  
- Este método ofrece una **reconstrucción más fiel** de la señal original que la detección por envolvente, ya que mantiene la información de fase.  
- Permite una recuperación más robusta ante interferencias o ruido, lo cual mejora la calidad de la señal demodulada.



## Aplicaciones de la Detección Coherente

###Radio AM y sistemas analógicos
En receptores AM síncronos o balanceados, la detección coherente permite obtener una señal de audio con menor distorsión y mejor calidad, gracias a la sincronización de la portadora local.

###Comunicaciones digitales modernas
Es esencial en modulaciones como **QAM**, **PSK** y **DSB-SC**, donde la información se transmite mediante variaciones de amplitud y/o fase.  
Esta técnica es la base de tecnologías como **Wi-Fi, televisión digital, 4G y 5G**, donde se requiere sincronización precisa entre transmisor y receptor.

###Comunicaciones inalámbricas y telemetría
Se emplea en sistemas donde la sincronización puede mantenerse estable (por ejemplo, enlaces satelitales, drones, y sensores remotos).  
Gracias a ello, se consigue **mayor estabilidad de transmisión** y **menor tasa de error**, incluso en entornos ruidosos.

###Procesamiento y análisis de señales
También se utiliza en el ámbito del procesamiento de señales para **extraer o medir componentes específicas en frecuencia** mediante multiplicación y filtrado.  
Se aplica en **radares, instrumentación industrial, mediciones de vibraciones y análisis de bioseñales**, donde permite detectar señales débiles en presencia de ruido o interferencias.


## Resumen

La **modulación por amplitud con detección coherente** es un método clásico pero aún muy relevante, ya que combina sencillez conceptual con precisión en la recuperación de la información. Su correcta sincronización entre transmisor y receptor garantiza una demodulación limpia y efectiva, siendo la base tanto de los sistemas analógicos tradicionales como de muchas tecnologías de comunicación digital modernas.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parámetros básicos ---
fc = 100              # Frecuencia de la portadora [Hz]
fm = 10               # Frecuencia del mensaje [Hz]
fs = 5 * fc           # Frecuencia de muestreo (suficiente para evitar aliasing)
t_final = 5 / fm      # Duración de la simulación
t = np.arange(0, t_final, 1/fs)

# --- Mensajes de entrada ---
# Señal tipo pulso rectangular
msg_pulso = np.zeros_like(t)
msg_pulso[(t >= t_final*0.4) & (t < t_final*0.6)] = 1

# Señal tipo coseno
msg_cos = np.cos(2 * np.pi * fm * t)

# --- Índice de modulación (usuario define) ---
m = float(input("Ingrese el índice de modulación (0 ≤ m ≤ 1): "))
A_pulso = np.max(np.abs(msg_pulso)) / m
A_cos = np.max(np.abs(msg_cos)) / m

# --- Portadora ---
carrier = np.sin(2 * np.pi * fc * t)

# --- Señales AM moduladas ---
am_pulso = (1 + msg_pulso / A_pulso) * carrier
am_cos = (1 + msg_cos / A_cos) * carrier
```

```python
# --- Función auxiliar para obtener el espectro ---
def espectro_fft(signal, fs):
    N = len(signal)
    f = np.fft.rfftfreq(N, 1/fs)
    X = np.fft.rfft(signal)
    return f, np.abs(X)

# --- Espectros en frecuencia ---
f_pulso, Y_pulso = espectro_fft(am_pulso, fs)
f_cos, Y_cos = espectro_fft(am_cos, fs)
f_msg_pulso, M_pulso = espectro_fft(msg_pulso, fs)
f_msg_cos, M_cos = espectro_fft(msg_cos, fs)
```

```python
# === Señal AM con mensaje tipo pulso ===
plt.figure(figsize=(8,4))
plt.plot(t, am_pulso, label='Señal AM (pulso)')
plt.plot(t, msg_pulso, '--', label='Mensaje pulso')
plt.title('Modulación AM - Mensaje tipo pulso')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()

# === Espectro del pulso y su modulación ===
plt.figure(figsize=(8,4))
plt.plot(f_pulso, Y_pulso, label='Espectro AM (pulso)')
plt.plot(f_msg_pulso, M_pulso, '--', label='Mensaje (pulso)')
plt.title('Espectro de frecuencia - Pulso rectangular')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('|X(f)|')
plt.legend()
plt.grid(True)
plt.show()

# === Señal AM con mensaje tipo coseno ===
plt.figure(figsize=(8,4))
plt.plot(t, am_cos, label='Señal AM (coseno)')
plt.plot(t, msg_cos, '--', label='Mensaje coseno')
plt.title('Modulación AM - Mensaje tipo coseno')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()

# === Espectro del coseno y su modulación ===
plt.figure(figsize=(8,4))
plt.plot(f_cos, Y_cos, label='Espectro AM (coseno)')
plt.plot(f_msg_cos, M_cos, '--', label='Mensaje (coseno)')
plt.title('Espectro de frecuencia - Señal cosenoidal')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('|X(f)|')
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_dem = s_t * c_t

f_d, D_f = obtener_espectro(y_dem, Fs)
plt.figure(figsize=(8,5))
plt.plot(f_d, D_f)
plt.title("Espectro de la señal demodulada (sin filtrar)")
plt.xlabel("Frecuencia [Hz]"); plt.ylabel("Magnitud"); plt.grid(True)
plt.xlim(0, 12000)
plt.show()
```

```python
X_fd = np.fft.rfft(y_dem)
freq_axis = np.fft.rfftfreq(len(y_dem), 1/Fs)
f_corte = 4000   # Hz
X_fd[freq_axis > f_corte] = 0

m_rec = np.fft.irfft(X_fd)
m_rec /= np.max(np.abs(m_rec))

plt.figure(figsize=(12,4))
plt.plot(t, m_t, '--', label="Mensaje original", alpha=0.7)
plt.plot(t, m_rec, color='red', label="Mensaje recuperado")
plt.title("Comparación: mensaje original vs. recuperado")
plt.xlabel("Tiempo [s]"); plt.ylabel("Amplitud"); plt.legend(); plt.grid(True)
plt.show()

print("Reproducción del mensaje recuperado:")
Audio(m_rec, rate=Fs)
```

## Distorsión Armónica Total (THD)

La **Distorsión Armónica Total** (*Total Harmonic Distortion*, THD) representa la cantidad de energía contenida en los armónicos (segundo, tercero, cuarto, etc.) de una señal periódica, en relación con la energía de su componente fundamental.

Si \( V_n \) es el valor eficaz (RMS) del n-ésimo armónico y \( V_1 \) el de la frecuencia fundamental, el THD se expresa como:

$$
THD = \frac{\sqrt{V_2^2 + V_3^2 + \dots + V_N^2}}{V_1}
$$

Frecuentemente se reporta en porcentaje:

$$
THD\% = 100 \times THD
$$

Un THD bajo indica que la señal se aproxima a una onda senoidal pura, mientras que un THD alto refleja una mayor presencia de distorsión armónica.


## Factor de Potencia (PF)

El **factor de potencia** mide qué tan eficientemente un circuito convierte la energía eléctrica suministrada en potencia útil. Es decir, evalúa la proporción entre la potencia activa y la potencia total entregada al sistema.

### • En sistemas senoidales ideales
Cuando el voltaje y la corriente son senoidales, el factor de potencia se calcula como:

$$
PF = \cos(\varphi)
$$

donde \( \varphi \) es el ángulo de desfase entre la tensión y la corriente.

- Si \( \varphi = 0 \), entonces \( PF = 1 \) → eficiencia máxima.  
- Si \( \varphi > 0 \), aparece potencia reactiva → \( PF < 1 \).

### • En sistemas no senoidales
Cuando existen **armónicos** en la forma de onda, el factor de potencia disminuye incluso sin desfase angular.  
En estos casos se define el **factor de potencia total considerando la distorsión (PFₜₕd):**

$$
PF_{THD} = \frac{1}{\sqrt{1 + (THD)^2}}
$$

Aquí, la reducción en el factor de potencia se debe a la energía distribuida en los armónicos en lugar de concentrarse en la fundamental.


## Cálculo del THD mediante FFT

El procedimiento general para calcular el THD de una señal digital muestreada es el siguiente:

1. Calcular la **FFT** (o `rFFT` si la señal es real) para obtener el espectro de frecuencias.
2. Trabajar con el **espectro de un solo lado** (solo frecuencias positivas) y normalizarlo dividiendo entre el número de muestras.
3. Identificar el **bin** correspondiente a la frecuencia fundamental \( f_1 \) y sus múltiplos \( 2f_1, 3f_1, \dots \).
4. Calcular \( V_n \) como la magnitud (en RMS) de cada armónico identificado.
5. Aplicar la fórmula del THD descrita arriba.

> En la práctica, se pueden detectar los armónicos buscando picos de amplitud en el espectro cerca de las frecuencias \( k f_1 \), con \( k = 1, 2, 3, \ldots \).


## Relación entre THD y Factor de Potencia

Cuando una señal presenta armónicos, parte de la energía no contribuye a la potencia activa, reduciendo así el factor de potencia efectivo.  
Una forma aproximada de cuantificar esta reducción es:

$$
PF_{THD} \approx \frac{1}{\sqrt{1 + THD^2}}
$$

Esta relación supone que el desfase entre la corriente y el voltaje fundamental es despreciable, por lo que la disminución en el factor de potencia proviene principalmente de la **distorsión armónica** y no del **desfase angular**.

```python
# Simulación: Cálculo del THD y factor de potencia en un
# rectificador de onda completa con carga resistiva y RC

import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
import sympy as sp

# 1. Definición simbólica del sistema RC
s = sp.symbols('s', complex=True)
R_, C_ = sp.symbols('R_ C_', positive=True)
Vin = sp.Function('Vin')(s)

# Función de transferencia RC (forma simbólica)
H_RC = (1 / (1 + R_ * C_ * s)) * Vin

# 2. Parámetros de simulación
f_linea = 60             # Frecuencia de la fuente (Hz)
fs = 40 * f_linea        # Frecuencia de muestreo (Hz)
periodo = 1 / f_linea
dt = 1 / fs
t_sim = np.arange(0, 5 * periodo, dt)   # Simulación de 5 ciclos
Vm = 120                # Amplitud de la fuente

# 3. Señal de entrada: onda senoidal rectificada completa
senal_seno = Vm * np.sin(2 * np.pi * f_linea * t_sim)
rect_full = np.abs(senal_seno)          # Rectificación completa

# 4. Función para simular salida según la carga
def salida_rect(R, C=None):
    """Genera la salida del rectificador según el tipo de carga."""
    if C is None:
        num, den = [1], [1]                  # Solo resistencia
    else:
        num, den = [1], [R * C, 1]           # Carga RC
    sistema = signal.TransferFunction(num, den)
    _, salida = signal.lsim(sistema, U=rect_full, T=t_sim)
    return salida

# 5. Cálculo del THD y del factor de potencia
def analizar_THD_PF(x, Fs):
    N = len(x)
    Xf = np.fft.rfft(x)
    freqs = np.fft.rfftfreq(N, 1 / Fs)
    mag = np.abs(Xf) / N

    # Encontrar la fundamental
    idx_fund = np.argmax(mag)
    V1 = mag[idx_fund]
    resto = np.delete(mag, idx_fund)
    THD = np.sqrt(np.sum(resto**2)) / V1
    PF = 1 / np.sqrt(1 + THD**2)
    return THD, PF, freqs, mag

# 6. Caso A: carga puramente resistiva
R_val = 1000
out_R = salida_rect(R_val)
THD_R, PF_R, fR, MagR = analizar_THD_PF(out_R, fs)

# 7. Caso B: carga RC (con capacidad en paralelo)
R_val = 1000
C_val = 22e-6
out_RC = salida_rect(R_val, C_val)
THD_RC, PF_RC, fRC, MagRC = analizar_THD_PF(out_RC, fs)

# 8. Resultados numéricos
print("===== RESULTADOS DEL RECTIFICADOR =====")
print(f"[Carga R pura]")
print(f"THD = {THD_R:.4f}  -> {THD_R*100:.2f}%")
print(f"Factor de potencia (THD): {PF_R:.4f}\n")

print(f"[Carga RC]")
print(f"THD = {THD_RC:.4f}  -> {THD_RC*100:.2f}%")
print(f"Factor de potencia (THD): {PF_RC:.4f}")
print("========================================\n")

# 9. Gráficas en el tiempo
plt.figure(figsize=(8,4))
plt.plot(t_sim, rect_full, label='Entrada rectificada', color='steelblue', lw=2.5)
plt.plot(t_sim, out_R, label='Salida (R)', color='crimson', lw=1.5)
plt.plot(t_sim, out_RC, label='Salida (RC)', color='seagreen', lw=2)
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.title('Rectificador de onda completa con diferentes cargas')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 10. Espectros de salida
plt.figure(figsize=(8,4))
plt.plot(fR, MagR, label='Carga R', lw=1.3)
plt.plot(fRC, MagRC, label='Carga RC', lw=1.3)
plt.xlim(0, 1000)
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud normalizada')
plt.title('Espectros de las salidas')
plt.legend()
plt.grid(alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Sistema 1: y[n] = x[n]/3 + 2 x[n-1] - y[n-1]
import numpy as np
from scipy.signal import lfilter

def shift_signal_causal(x, n0):
    N = len(x)
    if n0 == 0: return x.copy()
    if n0 > 0: return np.concatenate((np.zeros(n0), x))[:N]
    return np.concatenate((x[-n0:], np.zeros(-n0)))[:N]

# interpretar la ecuación como: y[n] + y[n-1] = (1/3)x[n] + 2 x[n-1]
b = [1/3.0, 2.0]   # para x[n], x[n-1]
a = [1.0, 1.0]     # para y[n], y[n-1]

# señales de prueba
N = 200
n = np.arange(N)
x1 = np.cos(0.2 * n)
x2 = np.sin(0.5 * n)

# parámetros para combinación
alpha, beta = 2.0, -3.0

# Linealidad (asumiendo y[-1]=0 porque lfilter inicializa en cero)
lhs = lfilter(b, a, alpha*x1 + beta*x2)
rhs = alpha * lfilter(b, a, x1) + beta * lfilter(b, a, x2)
is_linear = np.allclose(lhs, rhs, atol=1e-9)

# Invariancia en el tiempo (shift causal)
n0 = 7
lhs_ti = lfilter(b, a, shift_signal_causal(x1, n0))
rhs_ti = shift_signal_causal(lfilter(b, a, x1), n0)
is_ti = np.allclose(lhs_ti, rhs_ti, atol=1e-9)

print("Sistema 1 (suponer y[-1]=0):")
print("  Lineal?:", is_linear)
print("  Invariante en el tiempo?:", is_ti)
print("  SLIT?:", (is_linear and is_ti))
```

```python
# Sistema 2: y[n] = sum_{k=-inf}^{n} x[k]^2  (implementado como cumsum causal)
import numpy as np

def H2(x):
    return np.cumsum(x**2)

def shift_signal_causal(x, n0):
    N = len(x)
    if n0 == 0: return x.copy()
    if n0 > 0: return np.concatenate((np.zeros(n0), x))[:N]
    return np.concatenate((x[-n0:], np.zeros(-n0)))[:N]

# señales
n = np.arange(0, 50)
x = np.cos(0.2 * n)
alpha = 2.0

# Linealidad (homogeneidad)
lhs = H2(alpha * x)
rhs = alpha * H2(x)
is_linear = np.allclose(lhs, rhs, atol=1e-9)

# Invariancia en el tiempo
n0 = 5
lhs_ti = H2(shift_signal_causal(x, n0))
rhs_ti = shift_signal_causal(H2(x), n0)
is_ti = np.allclose(lhs_ti, rhs_ti, atol=1e-9)

print("Sistema 2:")
print("  Lineal?:", is_linear)
print("  Invariante en el tiempo?:", is_ti)
print("  SLIT?:", (is_linear and is_ti))
```

```python
# Sistema 3: mediana en ventana de 3
import numpy as np

def H3(x):
    N = len(x)
    y = np.zeros_like(x)
    for n in range(N):
        window = []
        for k in (n-1, n, n+1):
            if 0 <= k < N:
                window.append(x[k])
            else:
                window.append(0.0)
        y[n] = np.median(window)
    return y

def shift_signal_causal(x, n0):
    N = len(x)
    if n0 == 0: return x.copy()
    if n0 > 0: return np.concatenate((np.zeros(n0), x))[:N]
    return np.concatenate((x[-n0:], np.zeros(-n0)))[:N]

# contraejemplo para linealidad
x1 = np.array([0., 1., 10., 0., 0., 0.])
x2 = np.array([0., 10., 1., 0., 0., 0.])

lhs = H3(x1 + x2)
rhs = H3(x1) + H3(x2)
is_linear = np.allclose(lhs, rhs, atol=1e-9)

# invariancia en el tiempo
n0 = 2
lhs_ti = H3(shift_signal_causal(x1, n0))
rhs_ti = shift_signal_causal(H3(x1), n0)
is_ti = np.allclose(lhs_ti, rhs_ti, atol=1e-9)

print("Sistema 3:")
print("  Lineal?:", is_linear)
print("  Invariante en el tiempo?:", is_ti)
print("  SLIT?:", (is_linear and is_ti))
```

```python
# Sistema 4: y[n] = A x[n] + B
import numpy as np

def H4(x, A=2.0, B=5.0):
    return A * x + B

def shift_signal_causal(x, n0):
    N = len(x)
    if n0 == 0: return x.copy()
    if n0 > 0: return np.concatenate((np.zeros(n0), x))[:N]
    return np.concatenate((x[-n0:], np.zeros(-n0)))[:N]

# parámetros
A = 2.0
B = 0  # prueba con B != 0

# señales
N = 200
n = np.arange(N)
x1 = np.cos(0.05 * n)
x2 = np.sin(0.1 * n)

# Linealidad
lhs = H4(x1 + x2, A=A, B=B)
rhs = H4(x1, A=A, B=B) + H4(x2, A=A, B=B)
is_linear = np.allclose(lhs, rhs, atol=1e-9)

# Invariancia temporal
n0 = 10
lhs_ti = H4(shift_signal_causal(x1, n0), A=A, B=B)
rhs_ti = shift_signal_causal(H4(x1, A=A, B=B), n0)
is_ti = np.allclose(lhs_ti, rhs_ti, atol=1e-9)

print("Sistema 4 (A={}, B={}):".format(A, B))
print("  Lineal?:", is_linear, "(nota: será True solo si B==0)")
print("  Invariante en el tiempo?:", is_ti)
print("  SLIT?:", (is_linear and is_ti))
```

```python
# Ejemplo: Convolución discreta e interpretación paso a paso

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
from scipy import signal

# --------------------------------------------------------------
# 1. Definición de las señales de entrada y del sistema
# --------------------------------------------------------------

# Señal de entrada x[n]
x_n = np.array([-15, 5, -3, 0, 5, 7, -1]).reshape(-1, 1)

# Respuesta al escalón del sistema
u_resp = np.array([-1, 6, -10, 3, 1, -10, 2, 5])

# Respuesta al impulso obtenida como la diferencia entre muestras consecutivas
u_resp_ext = np.r_[0, u_resp, 0]   # se añade padding para obtener derivada discreta
h_n = np.diff(u_resp_ext).reshape(-1, 1)

# --------------------------------------------------------------
# 2. Función para graficar la convolución paso a paso
# --------------------------------------------------------------

def proceso_convolucion(h, x):
    """Crea las matrices necesarias para visualizar la convolución."""
    Lx, Lh = len(x), len(h)
    N = 2 * Lx + Lh
    x_mat = np.zeros((N, N))
    h_ext = np.r_[np.zeros((Lx, 1)), h, np.zeros((Lx, 1))]
    y_mat = np.zeros((N, N))

    for n in range(N - Lx + 1):
        x_mat[n:n + Lx, n] = np.flip(x).reshape(-1)
        y_mat[n, n] = x_mat[:, n].T @ h_ext  # producto punto

    return x_mat, y_mat, h_ext

def graficar_convolucion(indice, x_mat, y_mat, h_ext):
    """Grafica el proceso de convolución discreta para un índice dado."""
    plt.figure(figsize=(7, 3))
    plt.stem(x_mat[:, indice], markerfmt='o', linefmt='b', basefmt=' ')
    plt.stem(h_ext, linefmt='g', markerfmt='.', basefmt=' ')
    plt.stem(y_mat[:indice, :].sum(axis=0), linefmt='r', markerfmt='s', basefmt=' ')
    plt.title(r'Proceso de convolución $y[n] = \sum x[k]h[n-k]$')
    plt.legend([r'$x[k]$', r'$h[n-k]$', r'$y[n]$'])
    plt.grid(True, linestyle=':')
    plt.tight_layout()
    plt.show()

# --------------------------------------------------------------
# 3. Simulación interactiva del proceso de convolución
# --------------------------------------------------------------

x_m, y_m, h_ext = proceso_convolucion(h_n, x_n)

@interact(paso=(0, x_m.shape[0] - len(x_n), 1))
def mostrar_etapa(paso=0):
    graficar_convolucion(paso, x_m, y_m, h_ext)

# --------------------------------------------------------------
# 4. Cálculo directo de la salida por convolución
# --------------------------------------------------------------

y_num = np.convolve(x_n.reshape(-1), h_n.reshape(-1))

print("Secuencia de salida y[n]:")
print(y_num)

# --------------------------------------------------------------
# 5. Comparación opcional: Respuesta con otra h[n]
# --------------------------------------------------------------

# Si se desea comparar con la respuesta al impulso original:
h_alt = np.array([1, -2, 0, 1, -2]).reshape(-1, 1)
y_alt = np.convolve(x_n.reshape(-1), h_alt.reshape(-1))
print("\nSalida para h[n] alternativa = [1, -2, 0, 1, -2]:")
print(y_alt)
```
