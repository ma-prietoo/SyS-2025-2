```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
```

Punto 1

```python
t = np.linspace(-1, 5, 2000)
u = lambda x: np.where(x >= 0, 1, 0)

x1 = np.exp(-2*t)*u(t) + np.exp(-3*t)*u(t)

plt.figure(figsize=(8,4))
plt.plot(t, x1)
plt.title("Se√±al x1(t) = e^{-2t}u(t) + e^{-3t}u(t)")
plt.xlabel("t")
plt.ylabel("x1(t)")
plt.grid(True)
plt.show()
```

```python
num = [2, 5]          # 2s + 5
den = np.polynomial.polynomial.polyfromroots([-2, -3])[::-1]

system = signal.TransferFunction(num, den)

poles = system.poles
zeros = system.zeros

plt.figure(figsize=(5,5))
plt.scatter(poles.real, poles.imag, marker='x', s=150, label="Polos")
plt.scatter(zeros.real, zeros.imag, marker='o', s=150, label="Ceros")
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True)
plt.title("Polos y Ceros de X1(s)")
plt.legend()
plt.show()
```

Punto 2

```python
t = np.linspace(-5,5,3000)
u = lambda x: np.where(x>=0,1,0)

x2 = np.exp(2*t)*u(t) + np.exp(-3*t)*u(-t)

plt.figure(figsize=(8,4))
plt.plot(t, x2)
plt.title("Se√±al x2(t) = e^{2t}u(t) + e^{-3t}u(-t)")
plt.xlabel("t")
plt.ylabel("x2(t)")
plt.grid(True)
plt.show()
```

```python
num = [5]
den = np.polynomial.polynomial.polyfromroots([2, -3])[::-1]

system = signal.TransferFunction(num, den)

poles = system.poles
zeros = system.zeros

plt.figure(figsize=(5,5))
plt.scatter(poles.real, poles.imag, marker='x', s=150, label="Polos")
plt.scatter(zeros.real, zeros.imag, marker='o', s=150, label="Ceros")
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True)
plt.title("Polos y Ceros de X2(s) (ROC vac√≠a)")
plt.legend()
plt.show()
```

Punto 3

```python
a = 2   # Puedes cambiarlo

t = np.linspace(-5,5,3000)
x3 = np.exp(-a*np.abs(t))

plt.figure(figsize=(8,4))
plt.plot(t, x3)
plt.title("Se√±al x3(t) = e^{-a|t|}")
plt.xlabel("t")
plt.ylabel("x3(t)")
plt.grid(True)
plt.show()
```

```python
# X3(s) = 2a / (a^2 - s^2)

num = [2*a]
den = [1, 0, -(a**2)]   # s^2 - a^2 (solo cambia signo)

system = signal.TransferFunction(num, den)

poles = system.poles
zeros = system.zeros

plt.figure(figsize=(5,5))
plt.scatter(poles.real, poles.imag, marker='x', s=150, label="Polos")
plt.scatter(zeros.real, zeros.imag, marker='o', s=150, label="Ceros")
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True)
plt.title("Polos y Ceros de X3(s)")
plt.legend()
plt.show()
```

Punto 4

```python
t = np.linspace(-1, 7, 3000)
u = lambda x: np.where(x>=0,1,0)

x4 = np.exp(-2*t)*(u(t) - u(t-5))

plt.figure(figsize=(8,4))
plt.plot(t, x4)
plt.title("Se√±al x4(t) = e^{-2t}[u(t)-u(t-5)]")
plt.xlabel("t")
plt.ylabel("x4(t)")
plt.grid(True)
plt.show()
```

```python
# X4(s) = (1 - e^{-5(s+2)}) / (s+2)

# Polo √∫nico en s = -2
poles = np.array([-2])

# Ceros infinitos: s = -2 + j*(2*pi*k/5)
k = np.arange(-10, 11)
zeros = -2 + 1j*(2*np.pi*k/5)

plt.figure(figsize=(5,5))
plt.scatter(poles.real, poles.imag, marker='x', s=200, label="Polo")

plt.scatter(zeros.real, zeros.imag, marker='o', s=70, label="Ceros")

plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True)
plt.title("Polos y Ceros de X4(s)")
plt.legend()
plt.show()
```
