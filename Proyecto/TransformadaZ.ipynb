#Transformada de Laplace

La transformada de Laplace se puede entender como la transformada de Fourier de una funci칩n multiplicada por una exponencial decreciente, buscando cumplir las condici칩n d칠bil de Dirichlet $\int_T |x(t)|dt<\infty.$

### Transformada de Fourier de la se침al $x(t)$:

$$
X(\omega) = \int^{+\infty}_{-\infty}x(t)e^{-j\omega t}dt
$$

Si se m칰ltiplica la se침al $x(t)$ por la exponencial $e^{- \sigma t}$ y se le halla su transformada de Fourier, se obtiene la transformada de Laplace:

$$
\hat{x}(t,\sigma) = x(t)e^{-\sigma t}
$$

$$
\hat{X}(\sigma, \omega) = \int^{+\infty}_{-\infty}\hat{x}(t,\sigma)e^{-j\omega t}dt = \int^{+\infty}_{-\infty}x(t)e^{-\sigma t}e^{-j\omega t}dt
$$

Lo com칰n es escribir ambas variables en una sola, haciendo $\sigma + j\omega = s$:

$$
\hat{X}(\omega,\sigma) = X(s)
$$

$$
X(s) = \int^{+\infty}_{-\infty}x(t)e^{-s t}dt
$$

$$
\mathscr{L}\big\{ x(t)\big\}=\mathscr{F}\big\{ x(t)e^{-\sigma t}\big\}
$$



#Transformada Z
Para extender el concepto de transformada de Laplace a se침ales discretas, se modela una se침al $x_{m}(t)$ como una se침al continua $x(t)$ muestreada por un *ideal sampler* (tren de impulsos de periodo T):

$$
x_{m}(t) = \sum^{+\infty}_{n=-\infty}x(nT)\delta(t-nT)
$$

Obteniendo la transformada de Laplace de esta se침al:

$$
X_{m}(s) = \int^{+\infty}_{-\infty}\sum^{+\infty}_{n=-\infty}x(nT)\delta(t-nT)e^{-s t}dt
$$

$$
X_{m}(s) = \sum^{+\infty}_{n=-\infty}x(nT)\int^{+\infty}_{-\infty}\delta(t-nT)e^{-s t}dt = \sum^{+\infty}_{n=-\infty}x(nT)e^{-snT}
$$

Notando que $x(nT)$ es $x[n]$ y haciendo $z = e^{sT}$ se obtiene la transformada Z:

$$
X_{m}(s) = X(z)
$$

$$
X(z)= \sum^{+\infty}_{n=-\infty}x[n]z^{-n}
$$

#Relaci칩n entre el plano $s$ y el plano $z$
La relaci칩n entre el plano $s$ y el plano $z$ puede obtenerse expresando $s$ como n칰mero complejo:

$$
z = e^{sT}=e^{(\sigma+j\omega)T}
$$

Al expresar $e^{\sigma T}$ como $r$ y haciendo $\omega T =\Omega$ se obtiene la representaci칩n de $z$ en forma polar:

$$
z=re^{j\Omega}
$$

Es fundamental notar que los valores $r$ y $\Omega$ dependen del periodo de muestreo $T$.

Como $r$ depende de $\sigma$, l칤neas verticales en $s$ mapean circunferencias en $z$, mientras que l칤neas horizontales en $s$ ($\omega$ constante) mapean l칤neas rectas inclinadas en $z$.

![texto alternativo](https://lh4.googleusercontent.com/4xMYEQ5-XZOzDN6Up69rhFRmCMbZ1E9f0KXEDeRD5s4MeB9p3sJ07Jxzu8y2RxX4JGg45sFXdYUNz3WXFmN_beoeCG0_sPtsXp57sgI0L5K8qQ8xbTPEM1Z-K3xyoAbylnW1X1Ra)

- Recordar que la transformada de Fourier se obtiene cuando $\sigma =0$, es decir, evaluando $s$ sobre el eje vertical, quedando $s=j\omega$.

- De manera similar, debido a la dependencia de $r$ con respecto a $\sigma$, la transformada en tiempo discreto de Fourier (DTFT) se obtiene cuando $r=1$, es decir, evaluando $z$ sobre la circunferencia unitaria, quedando $z=e^{j\Omega}$.

Por este motivo, es que el espectro de una se침al discreta $x[n]$ se suele denotar como $X(e^{j\Omega}).$

#Modelado con la transformada Z

De manera similar, al modelado mediante ecuaciones diferenciales de los sistemas continuos, un SLIT discreto puede modelarse mediante ecuaciones de diferencias:

$$
a_{0}y[n]+a_{1}y[n-1]+a_{2}y[n-2]+\cdots + a_N y[n-N] = b_{0}x[n]+b_{1}x[n-1]+b_{2}x[n-2]+\cdots+b_M x[n-M]
$$

$$
\sum_{k=0}^{N}a_{k}y[n-k] = \sum_{k=0}^{M}b_{k}x[n-k]
$$

Aplicando la transformada $Z$ a esta expresi칩n:

$$
a_{0}Y(z)+a_{1}z^{-1}Y(z)+a_{2}z^{-2}Y(z)+\cdots+a_Nz^{-N}Y(z) = b_{0}X(z)+b_{1}z^{-1}X(z)+b_{2}z^{-2}X(z)+\cdots+b_Mz^{-M}X(z)
$$

**Nota**: Se sabe que $\mathfrak{Z}\{x[n\pm M]\}=z^{\pm M}X(z).$

La transformada Z plantea que:

$$
X(z)= \sum^{+\infty}_{n=-\infty}x[n]z^{-n}
$$

As칤:

$$\mathfrak{Z}\{x[n \pm M]\}= \sum^{+\infty}_{n=-\infty}x[n \pm M]z^{-n}$$

Para $n'=n\pm M;$  $n=n'\mp M:$

$$\sum^{+\infty}_{n'=-\infty}x[n']z^{-(n'\mp M)}=\sum^{+\infty}_{n'=-\infty}x[n']z^{-n'}z^{\pm M}$$

$$z^{\pm M}\sum^{+\infty}_{n'=-\infty}x[n']z^{-n'}=z^{\pm M}X(z)$$

Por ende,

$$
\bigg(\sum_{k=0}^{N}a_{k}z^{-k}\bigg)Y(z)= \bigg(\sum_{k=0}^{M}b_{k}z^{-k}\bigg)X(z)
$$

Al hallar la funci칩n de transferencia se observa que un SLIT discreto puede modelarse como un cociente de polinomios en $z$
:

$$H(z) = \frac{Y(z)}{X(z)} = \frac{\sum_{k=0}^{M}b_{k}z^{-k}}{\sum_{k=0}^{N}a_{k}z^{-k}}
$$

# Sistemas no recursivos (FIR)

Un sistema que carece de realimentacion se denomina sistema de *respuesta finita al impulso* (FIR). Al carecer de realimentaci칩n, se tiene que $a_{k}=0, \forall k=\{1, 2, 3, ..., N\}$, y $a_{0}=1$. De modo que su funci칩n de transferencia es (solo ceros):

$$
H(z) = \frac{Y(z)}{X(z)} = \sum_{k=0}^{M}b_{k}z^{-k}
$$

# Sistemas recursivos (IIR)

La forma can칩nica directa de los sistemas de *respuesta infinita al impulso* (IIR) se obtiene haciendo $a_{0}=1$, con lo que su funci칩n de transferencia se define como:

$$
H(z) = \frac{Y(z)}{X(z)} = \frac{\sum_{k=0}^{M}b_{k}z^{-k}}{1-\sum_{k=1}^{N}a_{k}z^{-k}}
$$

#Dise침o de Filtros IIR

##Invarianza del impulso
Este m칠todo de dise침o se basa en el muestreo de la respuesta impulso de un filtro en tiempo continuo $h_{c}(t)$, multiplicada por el per칤odo de muestreo $T_{s}$:

$$
h[n]=T_{s}h_{c}(nT_{s}).
$$

Un sistema continuo se puede representar en $s$ como una suma de fracciones parciales:

$$
H_{c}(s) = \sum_{k=1}^{N}\frac{A_{k}}{s-s_{k}}.
$$

$$\frac{Y(s)}{X(s)}=H(s),$$ $x(t)=\delta(t)$

$X(s)=1$

La correspondiente funci칩n en tiempo es entonces (Ver [tablas transformada de Laplace](https://tutorial.math.lamar.edu/classes/de/laplace_table.aspx)):

$$ h_{c}(t)=
  \begin{cases}
    \sum_{k=1}^{N}A_{k}e^{s_{k}t}       & \quad t\geq0\\
    0  & \quad t<0
  \end{cases}
$$

Realizando el muestreo:

$$
h[n] = T_{s}\sum_{k=1}^{N}A_{k}e^{s_{k}nT_{s}}u[n].
$$

Al obtener la transformada $Z$ (Ver [tablas transformada Z](https://lpsa.swarthmore.edu/LaplaceZTable/LaplaceZFuncTable.html)), se obtiene:

$$
H(z)=\sum_{k=1}^{N}\frac{T_{s}A_{k}}{1-e^{s_{k}T_{s}}z^{-1}}
$$

De manera que conociendo los polos $s=s_{k}$ se puede obtener los polos $z=e^{s_{k}T_{s}}$.



###Ejemplo de dise침o de un filtro en tiempo discreto a partir de su contraparte en tiempo continuo

Sup칩ngase que se tiene un filtro con los siguientes polos, queriendo realizar un filtrado con $T_{s}=1$:

$$
s_{1,2}=-\frac{1}{2}\pm j\frac{1}{2}
$$

La correspondiente funci칩n de transferencia en $s$ se encuentra seg칰n sus polos como:

$$
H_{c}(s) = \frac{j}{s+\frac{1+j}{2}} - \frac{j}{s+\frac{1-j}{2}}
$$

Los polos en $z$ se calculan de la manera indicada anteriormente

$$
z_1=e^{-\frac{1+j}{2}}
$$
$$
z_2=e^{-\frac{1-j}{2}}
$$

Finalmente, obteniendo la funci칩n de transferencia en $z$

$$
H(z)=\frac{j}{1-e^{-\frac{1+j}{2}}z^{-1}} -\frac{j}{1-e^{-\frac{1-j}{2}}z^{-1}}
$$

$$
H(z)=\frac{jz}{z-e^{-\frac{1+j}{2}}} -\frac{jz}{z-e^{-\frac{1-j}{2}}}
$$

##Transformaci칩n bilineal

Este m칠todo consiste en realizar el siguiente mapeo (Ver aproximaci칩n funci칩n logartimo en [transformaci칩n bilineal](https://en.wikipedia.org/wiki/Bilinear_transform))

$$
s=\frac{2}{T_{s}}\bigg(\frac{1-z^{-1}}{1+z^{-1}}\bigg)
$$

Al evaluar $H_{c}(s)$ en tal expresi칩n, se obtiene  la funci칩n de transferencia en $z$

$$
H(z)=H_{c}\Bigg(\frac{2}{T_{s}}\bigg(\frac{1-z^{-1}}{1+z^{-1}}\bigg)\Bigg)
$$

El mapeo correspondiente de $s$ a $z$ se obtiene despejando $z$ de la primera expresi칩n

$$
z=\frac{1+(T_{s}/2)s}{1-(T_{s}/2)s}
$$

###Ejemplo
Supongase que se tiene el mismo filtro anterior, pero esta vez se desea obtener la expresi칩n en $z$ mediante transformaci칩n bilinear. Realizando la evaluaci칩n de $s$ como se expuso

$$
H_{c}\bigg(2\Big(\frac{1-z^{-1}}{1+z^{-1}}\Big)\bigg) = \frac{j}{2\Big(\frac{1-z^{-1}}{1+z^{-1}}\Big)+\frac{1+j}{2}} - \frac{j}{2\Big(\frac{1-z^{-1}}{1+z^{-1}}\Big)+\frac{1-j}{2}}
$$

Desarrollando algrebraicamente se llega finalmente a la expresi칩n

$$
H(z)=\frac{1+5j}{13}\cdot\frac{z+1}{z+\frac{-7+4j}{13}} +\frac{1-5j}{13}\cdot\frac{z+1}{z-\frac{7+4j}{13}}
$$

Que es precisamente la funci칩n de transferencia que se quer칤a encontrar.

Este 칰ltimo es el m칠todo que se suele usar para dise침ar aproximaciones de los filtros Butterworth, Chebyshev y el칤pticos en tiempo discreto.

# Filtrado y el retardo de fase

- Dado que el filtro digital se puede entender como un sistema SLIT, su respuesta al impulso genera un desfase en la frecuencia que se evidencia como un retardo en el tiempo. El objetivo es mantener dicho retardo estable y conocido para compensarlo.



$$\mathfrak{F}\{x(t\pm t_0)\}=X(w)e^{\mp j w t_0}$$

$$X(w)=|X(w)|e^{j\theta}$$

$$y(t)=\mathfrak{H}\{x(t)\}=h(t)*x(t)$$

$$Y(w)=X(w)H(w)=|X(w)||H(w)|e^{j(\theta_x+\theta_h)}$$


- Algunas alternativas para compensar el retarda en fase incluyen la estimaci칩n del mismo desde los puntos en el tiempo (ver [lfitler_zi](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.lfilter_zi.html))) o la evaluaci칩n del filtro hacia adelante y hacia atr치s (ver [filtfilt](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.filtfilt.html)).



### FIltro Butterworth
El [filtro Butterworth](https://en.wikipedia.org/wiki/Butterworth_filter) posee la ventaja de contar con la respuesta m치s plana posible para la banda pasante, sin embargo la banda de transici칩n es muy amplia.

```python
from scipy.signal import butter as bw
from scipy.signal import freqz_zpk
```

```python
N = 1 #orden el filtro
Wn = [2000, 5000] #banda pasante
filt = 'bandpass' #tipo del filtro (pasa-baja, pasa-alta, pasabanda, rechazabanda)
out = 'zpk' #la funcion retorna los polos y ceros junto con la ganancia. Para usar numerador y denominador usar 'ba' junto con freqz
zeros, poles, gain = bw(N, Wn, btype=filt, output=out, fs=fs)
#K = len(Xw[:,0]) #longitud del vector transformada
w, h = freqz_zpk(zeros, poles, gain, fs=fs) #Para indicar la longitud usar worN=K
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.patches import Circle

def plot_freq_response(filter_name, w, h, N): #Mostrar la respuesta en frecuencia del filtro
  fig = plt.figure()
  ax1 = fig.add_subplot(1, 1, 1)
  ax1.set_title(filter_name +' Digital filter frequency response, order= ' + str(N))
  ax1.plot(w, 20 * np.log10(abs(h)), 'b')
  ax1.set_ylabel('Amplitude [dB]', color='b')
  ax1.set_xlabel('Frequency [Hz]')
  ax1.grid()
  ax2 = ax1.twinx()
  angles = np.unwrap(np.angle(h))
  ax2.plot(w, angles, 'g')
  ax2.set_ylabel('Angle [radians]', color='g')
  plt.axis('tight')
  plt.show()
```

```python
def show_zp(z, p, title= 'Z-plane'): # Mostrar la ubicaci칩n de los polos y los zeros
  ax = plt.gca()

  ax.plot(np.real(z), np.imag(z), 'bo', fillstyle='none', ms = 10)
  ax.plot(np.real(p), np.imag(p), 'rx', fillstyle='none', ms = 10)
  unit_circle = Circle((0,0), radius=1, fill=False,
                        color='black', ls='solid', alpha=0.9)
  ax.add_patch(unit_circle)
  ax.axvline(0, color='0.7')
  ax.axhline(0, color='0.7')

  plt.title(title)
  plt.xlabel(r'Re{$z$}')
  plt.ylabel(r'Im{$z$}')
  plt.axis('equal')
  plt.xlim((-2, 2))
  plt.ylim((-2, 2))
  plt.grid()
```

```python
show_zp(zeros, poles, title='Polos y Ceros')
```

```python
plot_freq_response('Butterworth', w, h, N)
```

```python
#Puede obtenerse un mejor resultado aumentando el grado del filtro, a cambio de un mayor coste computacional y problemas en la fase
N = 4
zeros, poles, gain = bw(N, Wn, btype=filt, output=out, fs=fs)
w, h = freqz_zpk(zeros, poles, gain, fs=fs)

plot_freq_response('Butterworth', w, h, N)
```

### Filtro Chebyshev
Al contrario del filtro Butterworth, el filtro de [Chebyshev](https://en.wikipedia.org/wiki/Chebyshev_filter) tiene una pendiente en la ca칤da muy pronunciada, por lo que la banda de transici칩n resulta ser menor, a cambio de que existe un rizado en la banda pasante y/o rechazada. Dependiendo del lugar en el que se escoja el rizado, se tienen dos filtros diferentes de Chebyshev.


#### Chebyshev 1
Para este filtro se busca un mayor rechazo en la banda rechazada a cambio de obtener un rizado en la banda pasante

```python
from scipy.signal import cheby1
```

```python
N = 4
ripple = 10
zeros, poles, gain = cheby1(N, rp=ripple, Wn=Wn, btype=filt, output=out, fs=fs)
w, h = freqz_zpk(zeros, poles, gain, fs=fs)
```

```python
show_zp(zeros, poles)
```

```python
plot_freq_response('Chebyshev1', w, h, N)
```

#### Chebyshev 2
Para este caso del filtro Chebyshev, se prefiere que se tenga una respuesta plana en la banda pasante, obteniendo ahora el rizo en la banda rechazada

```python
from scipy.signal import cheby2
from scipy.signal import freqz #para este caso no se van a mostrar los polos y ceros
```

```python
N = 4
ripple = 20
out = 'ba' #no regresa los polos y ceros, sino la funcion de transferencia
num, den = cheby2(N, rs=ripple, Wn=Wn, btype=filt, output=out, fs=fs) #regresa numerador y denominador. Si se quiere obtener los polos y ceros usar scipy.signal.tf2zpk
w, h = freqz(num, den, fs=fs)
```

```python
plot_freq_response('Chebyshev2', w, h, N)
```

### Filtro El칤ptico (Filtro de Cauer)
Este filtro [El칤ptico](https://es.wikipedia.org/wiki/Filtro_de_Cauer#:~:text=Un%20filtro%20el%C3%ADptico%20o%20filtro,redes%20y%20dise%C3%B1o%20de%20filtros) es el que contiene la banda de transici칩n m치s delgada posible, incluyendo adem치s rizado tanto en la banda pasante como en la rechazada (ambos rizados son independientes entre s칤)

```python
from scipy.signal import ellip
```

```python
N = 4
ripple_pass = 2 #para la banda pasante
ripple_stop = 30 #para la banda rechazada
num, den = ellip(N, rp=ripple_pass, rs=ripple_stop, Wn=Wn, btype=filt, output=out, fs=fs) #regresa numerador y denominador. Si se quiere obtener los polos y ceros usar scipy.signal.tf2zpk
w, h = freqz(num, den, fs=fs)
```

```python
plot_freq_response('Elliptic', w, h, N)
```

### Filtro de Bessel
El filtro de [Bessel](https://es.wikipedia.org/wiki/Filtro_de_Bessel)
posee una respuesta casi lineal en fase.


```python
from scipy.signal import bessel
```

```python
N = 4
normalization = 'mag' #Ajusta la frecuencia cr칤tica seg칰n la respuesta que se desea
num, den = bessel(N, Wn=Wn, btype=filt, norm=normalization, output=out, fs=fs) #regresa numerador y denominador. Si se quiere obtener los polos y ceros usar scipy.signal.tf2zpk
w, h = freqz(num, den, fs=fs)
```

```python
plot_freq_response('Bessel', w, h, N)
```

## Filtro paso alto
En los instrumentos musicales existe una banda de frecuencias en los que se pueden tocar, mientras que en otros les es imposible. Por este motivo es que es posible suprimir algunos instrumentos de una pista mediante el uso de filtros.

Como ejemplo, supongase que en el audio anterior se quiere suprimir el sonido del bajo el칠ctrico. Como su nombre lo indica, este es un instrumento que se caracteriza por manejar una banda de frecuencias bajas (sonidos graves), por lo que aplicando un filtro de paso alto se podr칤a realizar esta tarea.



```python
filtro = "highpass"
N = 6
frec = 1000

ripple_pass = 2
ripple_stop = 101
num, den = ellip(N, rp=ripple_pass, rs=ripple_stop, Wn=frec, btype=filtro, fs=fs)
```

```python
from scipy.signal import lfilter, filtfilt # filtfilt compensa el retardo en tiempo debido a la respuesta en fase del filtro
```

```python
xf = filtfilt(num, den, xs, axis=0)
```

```python
Audio([xf[:,1], xf[:,0]],rate=fs) #Resultado
```

## Filtro paso bajo
Un problema que le ocurre continuamente a las se침ales es la presencia de ruido. A continuaci칩n se muestra el mismo audio contaminado con ruido blanco

```python
x_n = xs + np.random.normal(loc=0, scale=0.05, size=xs.shape)
```

```python
Audio([x_n[:,1], x_n[:,0]],rate=fs) #Resultado
```

Por lo general el ruido se caracteriza por manejar una banda de frecuencias altas. Teniendo esto en cuenta, puede deshacerse de dicho ruido aplicando un filtro que suprima las frecuencias altas, es decir, un filtro pasabajas.

```python
filtro = "lowpass"
N = 10
frec = 1600

ripple_pass = 1
ripple_stop = 100
num, den = ellip(N, rp=ripple_pass, rs=ripple_stop, Wn=frec, btype=filtro, fs=fs)

xf = lfilter(num, den, x_n, axis=0)
```

En el audio filtrado se percibe menos ruido, pero las fecuencias altas del audio original tambi칠n se ven afectadas.

```python
Audio([xf[:,1], xf[:,0]],rate=fs) #Audio con ruido filtrado
```

```python
import os
import numpy as np
import soundfile as sf
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from scipy.signal import (
    butter, cheby1, cheby2, ellip, bessel,
    freqz, freqz_zpk, filtfilt, lfilter,
    firwin, tf2zpk
)
from IPython.display import Audio
import pandas as pd
```

```python
# Funciones para ejecutar comandos shell
def run_cmd(cmd):
    print("Running:", cmd)
    result = subprocess.run(shlex.split(cmd), capture_output=True, text=True)
    if result.stdout:
        print(result.stdout)
    if result.stderr:
        print(result.stderr)
    if result.returncode != 0:
        raise RuntimeError(f"Command failed: {cmd}")

# Descargar segmento de YouTube y convertir a wav recortado
def download_youtube_segment(youtube_link, out_wav="segment.wav", start=0, duration=10, tmp_prefix="temp"):
    # descarga mp3 (yt-dlp)
    cmd_dl = f'yt-dlp --extract-audio --audio-format mp3 -o "{tmp_prefix}.%(ext)s" "{youtube_link}"'
    run_cmd(cmd_dl)
    tmp_mp3 = tmp_prefix + ".mp3"
    # a veces el nombre no es exactamente temp.mp3 -> buscar
    if not os.path.exists(tmp_mp3):
        candidates = [f for f in os.listdir(".") if f.startswith(tmp_prefix) and f.endswith(".mp3")]
        if candidates:
            tmp_mp3 = candidates[0]
        else:
            raise FileNotFoundError("No se encontr칩 el mp3 descargado por yt-dlp.")
    # convertir a wav y recortar con ffmpeg
    ss = f"-ss {float(start)}" if start is not None else ""
    cmd_ffmpeg = f'ffmpeg -y {ss} -t {float(duration)} -i "{tmp_mp3}" -ar 44100 -ac 2 "{out_wav}"'
    run_cmd(cmd_ffmpeg)
    print("Segmento guardado en:", out_wav)
    return out_wav

def read_wav(path):
    x, fs = sf.read(path)
    return x, fs

def save_wav(path, data, fs):
    sf.write(path, data, fs)
    print("Guardado:", path)
```

```python
# Dise침o IIR seg칰n tipo
def design_iir(filter_name, order, Wn, fs, **kwargs):
    out = 'ba'  # numerador/denominador
    btype = kwargs.get('btype', 'bandpass')
    if filter_name == 'butter':
        num, den = butter(order, Wn, btype=btype, fs=fs, output=out)
    elif filter_name == 'cheby1':
        num, den = cheby1(order, rp=kwargs.get('rp',1), Wn=Wn, btype=btype, fs=fs, output=out)
    elif filter_name == 'cheby2':
        num, den = cheby2(order, rs=kwargs.get('rs',40), Wn=Wn, btype=btype, fs=fs, output=out)
    elif filter_name == 'ellip':
        num, den = ellip(order, rp=kwargs.get('rp',1), rs=kwargs.get('rs',40), Wn=Wn, btype=btype, fs=fs, output=out)
    elif filter_name == 'bessel':
        num, den = bessel(order, Wn, btype=btype, fs=fs, output=out, norm='mag')
    else:
        raise ValueError("Filtro desconocido")
    return num, den

def apply_filter_zero_phase(num, den, x):
    # intentamos filtfilt para evitar retardo de fase; si falla, usamos lfilter
    try:
        y = filtfilt(num, den, x, axis=0)
    except Exception as e:
        print("filtfilt fall칩:", e, "-> usando lfilter")
        y = lfilter(num, den, x, axis=0)
    return y
```

#Dise침o de Filtros FIR por Ventaneo

**1. Concepto general del m칠todo de ventaneo**

El m칠todo de ventaneo consiste en aproximar la respuesta al impulso ideal de un filtro deseado mediante el uso de una ventana de longitud finita.  
Un filtro FIR de orden \(M\) tiene la forma:
$$
H(z) = \sum_{k=0}^{M} b_k z^{-k}
$$

Es decir, solo tiene ceros, pues en un FIR:

$$
a_0 = 1, \qquad a_k = 0 \quad \forall k \ge 1.
$$

El procedimiento general es:

1. Definir la **respuesta en frecuencia ideal** deseada \(H_d(e^{j\omega})\).  
2. Obtener la **respuesta al impulso ideal** mediante transformada inversa:
   $$
   h_d[n] = \frac{1}{2\pi} \int_{-\pi}^{\pi} H_d(e^{j\omega}) e^{j\omega n}\, d\omega.
   $$
3. Como h_d[n] suele ser infinita, se **trunca y pondera** con una ventana:
   $$
   h[n] = h_d[n] \, w[n], \qquad 0 \le n \le M.
   $$
4. El resultado es un filtro FIR **causal, estable y de fase lineal** (si se asigna simetr칤a).

**2. Respuesta ideal para un filtro paso bajo**

Para un filtro paso bajo con frecuencia de corte normalizada \(\omega_c\), la respuesta ideal es:

$$
H_d(e^{j\omega}) =
\begin{cases}
1, & |\omega| \le \omega_c \\
0, & \text{caso contrario}
\end{cases}
$$
La respuesta al impulso ideal correspondiente (tipo sinc desplazada) es:

$$
h_d[n] =
\begin{cases}
\frac{\sin\left(\omega_c (n - M/2)\right)}{\pi (n - M/2)}, & n \ne M/2\\[6pt]
\frac{\omega_c}{\pi}, & n = M/2
\end{cases}
$$

---

**3. Aplicaci칩n del ventaneo**

Un FIR se obtiene multiplicando la sinc ideal por una ventana \(w[n]\):

$$
h[n] = h_d[n]\; w[n]
$$

Las ventanas m치s comunes son:

- **Rectangular**  
  $$
  w[n] = 1
  $$
  Mejor transici칩n, peor atenuaci칩n.

- **Hann**  
  $$
  w[n] = 0.5 - 0.5\cos\left(\frac{2\pi n}{M}\right)
  $$

- **Hamming**  
  $$
  w[n] = 0.54 - 0.46\cos\left(\frac{2\pi n}{M}\right)
  $$

- **Blackman**  
  $$
  w[n] = 0.42 - 0.5\cos\left(\frac{2\pi n}{M}\right) + 0.08\cos\left(\frac{4\pi n}{M}\right)
  $$

- **Kaiser**, controlable mediante \(\beta\)  
  $$
  w[n] = \frac{I_0\left(\beta \sqrt{1 - \left(\frac{2n}{M} - 1\right)^2}\right)}
  {I_0(\beta)}
  $$

---

**4. C치lculo del orden con ventana de Kaiser**

Dado un rizado m치ximo \(\delta\) o atenuaci칩n:

$$
A = -20 \log_{10}(\delta)
$$

y un ancho de transici칩n normalizado \(\Delta \omega\), el orden aproximado es:

$$
N \approx \frac{A - 8}{2.285 \, \Delta\omega}
$$

El par치metro \(\beta\) depende de \(A\):

$$
\beta =
\begin{cases}
0.1102 (A - 8.7), & A > 50 \\
0.5842(A - 21)^{0.4} + 0.07886(A - 21), & 21 \le A \le 50 \\
0, & A < 21
\end{cases}
$$

---

## **5. Ventajas y desventajas de FIR e IIR**

**Cuadro comparativo**

| Caracter칤stica | FIR | IIR |
|----------------|-----|-----|
| Polos | No tiene (solo ceros) | Tiene polos y ceros |
| Estabilidad | Siempre estable | Puede ser inestable |
| Fase | Puede ser **lineal exacta** | Generalmente **no lineal** |
| Orden requerido | Alto | Bajo |
| Complejidad computacional | Alta | Baja |
| Retardo | Alto (\(M/2\)) | Bajo |
| Robustez num칠rica | Muy buena | Sensible |
| Dise침o | Muy simple (ventaneo, sinc) | M치s complejo |
| Aplicaciones | Audio, comunicaciones, multirate, sistemas cr칤ticos | Control, sistemas embebidos, filtrado eficiente |

---

**6. Dise침o paso a paso (resumen pr치ctico)**

1. Especificar:  
   $$
   f_s,\quad f_p,\quad f_s,\quad A_{pass},\quad A_{stop}
   $$
2. Calcular \(\Delta f\) y normalizar:  
   $$
   \Delta f_{norm} = \frac{f_s - f_p}{f_{Nyquist}}
   $$
3. Seleccionar ventana (Hamming, Blackman, Kaiser).  
4. Si es Kaiser, calcular orden:  
   $$
   N \approx \frac{A-8}{2.285\Delta\omega}
   $$
5. Generar:  
   $$
   h[n] = h_d[n]\, w[n]
   $$
6. Evaluar respuesta en frecuencia y ajustar.

\end{cases}

El dise침o FIR por ventaneo es un m칠todo simple, robusto y efectivo cuando se busca:

## **Filtros FIR e IIR y su aplicaci칩n en se침ales EEG para tareas de Imaginaci칩n Motora (MI)**

En el procesamiento digital de se침ales, y particularmente en el an치lisis de se침ales EEG, los filtros digitales constituyen una herramienta fundamental para extraer informaci칩n relevante y reducir componentes no deseados. A partir del estudio de la Transformada Z, los sistemas discretos se modelan mediante ecuaciones en diferencias y funciones de transferencia en el dominio *z*, lo que permite dise침ar y analizar filtros FIR e IIR de manera rigurosa.

### **Filtros FIR (Finite Impulse Response)**

Los filtros FIR se caracterizan porque su respuesta al impulso es finita, es decir, el sistema solo depende de un n칰mero limitado de muestras pasadas de la se침al de entrada. Matem치ticamente, se describen como:

$
y[n] = \sum_{k=0}^{M} b_k \, x[n-k]
$

En el dominio de la Transformada Z, su funci칩n de transferencia es un polinomio en \( z^{-1} \), sin polos distintos del origen, lo que implica que:

- Son siempre estables.
- Pueden dise침arse con fase lineal, una propiedad muy valiosa cuando se desea preservar la forma temporal de las se침ales EEG.
- Su implementaci칩n es conceptualmente simple, aunque suelen requerir un orden alto para lograr transiciones espectrales pronunciadas.

#### **Uso de filtros FIR en EEG y MI**
En tareas de Imaginaci칩n Motora, los filtros FIR se emplean principalmente para:
- **Resaltar patrones espectrales**, como las bandas 풮 (813 Hz) y  (1330 Hz), asociadas a la activaci칩n motora cortical.
- **Preservar la informaci칩n temporal**, evitando distorsiones de fase que podr칤an alterar la detecci칩n de eventos relacionados con ERD/ERS (Event-Related Desynchronization/Synchronization).
- Servir como etapa previa en m칠todos de an치lisis espacial, asegurando se침ales limpias y coherentes entre canales.

### **Filtros IIR (Infinite Impulse Response)**

Los filtros IIR presentan una respuesta al impulso infinita, ya que utilizan realimentaci칩n y dependen tanto de muestras pasadas de la entrada como de la salida:

$
y[n] = \sum_{k=0}^{M} b_k \, x[n-k] - \sum_{k=1}^{N} a_k \, y[n-k]
$

En el dominio z, su funci칩n de transferencia es una funci칩n racional, con polos y ceros que determinan su comportamiento din치mico.

Las principales caracter칤sticas de los filtros IIR son:
- Mayor **eficiencia computacional**, logrando respuestas selectivas con 칩rdenes bajos.
- Capacidad de aproximar filtros anal칩gicos cl치sicos (Butterworth, Chebyshev, El칤ptico, Bessel) mediante t칠cnicas como **invariancia al impulso** o **transformaci칩n bilineal**.
- Posible **distorsi칩n de fase**, lo cual debe considerarse cuidadosamente en aplicaciones neurofisiol칩gicas.

#### **Uso de filtros IIR en EEG y MI**
En el contexto de EEG y MI, los filtros IIR se utilizan para:
- **Aislar r치pidamente bandas de frecuencia** relevantes con bajo costo computacional.
- Realizar **preprocesamiento en tiempo real**, donde la latencia y la complejidad son cr칤ticas.
- Eliminar artefactos de baja o alta frecuencia (deriva DC, ruido muscular, interferencias).

### **Relaci칩n con patrones espaciales, temporales y espectrales en MI**

El an치lisis de se침ales EEG en tareas de Imaginaci칩n Motora requiere identificar patrones en distintos dominios:

- **Patrones espectrales**:  
  Los filtros FIR e IIR permiten seleccionar bandas espec칤ficas (풮 y ), donde se manifiestan cambios de potencia relacionados con la actividad motora imaginada.

- **Patrones temporales**:  
  Los filtros FIR con fase lineal preservan la din치mica temporal de la se침al, facilitando el an치lisis de eventos transitorios como ERD/ERS durante la ejecuci칩n de la tarea MI.

- **Patrones espaciales**:  
  Aunque los filtros digitales act칰an sobre cada canal individualmente, su aplicaci칩n previa es esencial para t칠cnicas espaciales como **Common Spatial Patterns (CSP)**, ya que mejora la relaci칩n se침alruido y realza la coherencia entre canales corticales relevantes.

### **Conclusi칩n**

En s칤ntesis, los filtros FIR e IIR, dise침ados y analizados mediante la Transformada Z, son herramientas complementarias en el procesamiento de se침ales EEG para Imaginaci칩n Motora.  

Los FIR destacan por su estabilidad y preservaci칩n temporal, mientras que los IIR ofrecen eficiencia y selectividad espectral. Su correcta aplicaci칩n permite resaltar patrones temporales, espectrales y espaciales fundamentales para el desarrollo de sistemas BCI robustos y confiables.
