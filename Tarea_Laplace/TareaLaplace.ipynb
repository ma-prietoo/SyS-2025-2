##Ejercicio 2

```python
import numpy as np
import matplotlib.pyplot as plt

# ===============================
# Parmetros del circuito
# ===============================
R = 1000            # Ohm
L = 180e-3          # Henry
C = 120e-6          # Farad

# ===============================
# Vector de frecuencias
# ===============================
w = np.logspace(0, 5, 3000)   # de 1 a 100000 rad/s

# ===============================
# Clculo de la funcin H(jw)
# ===============================
A = 1 - L*C*w**2        # parte real
B = R*C*w               # parte imaginaria

H = 1 / (A + 1j*B)      # funcin compleja

# ===============================
# Magnitud en dB
# ===============================
mag = 20 * np.log10(np.abs(H))

# ===============================
# Fase en grados
# ===============================
phase = np.angle(H, deg=True)

# ===============================
# Grfico de magnitud
# ===============================
plt.figure()
plt.semilogx(w, mag)
plt.xlabel(" (rad/s)")
plt.ylabel("Magnitud (dB)")
plt.title("Diagrama de Bode - Magnitud")
plt.grid(True)
plt.show()

# ===============================
# Grfico de fase
# ===============================
plt.figure()
plt.semilogx(w, phase)
plt.xlabel(" (rad/s)")
plt.ylabel("Fase ()")
plt.title("Diagrama de Bode - Fase")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Polos del ejercicio (VALORES CORRECTOS)
# L = 180 mH, C = 120 uF, R = 1000 ohm
p1 = -9.3519 + 0j
p2 = -5545.2990 + 0j

poles = np.array([p1, p2])

# No hay ceros en este sistema (numerador = 1)
zeros = np.array([])

plt.figure(figsize=(8,8))

# Graficar polos
plt.scatter(np.real(poles), np.imag(poles),
            marker='x', s=120, color='red', label='Polos')

# Graficar ceros si hubiera (no hay)
if len(zeros) > 0:
    plt.scatter(np.real(zeros), np.imag(zeros),
                marker='o', facecolors='none', edgecolors='blue',
                s=120, label='Ceros')

# Ejes
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

plt.title("Diagrama de Polos y Ceros")
plt.xlabel("Parte Real ()")
plt.ylabel("Parte Imaginaria ()")
plt.grid(True, which='both')
plt.legend()
plt.show()
```
